```python
#import pandas as pd
#import numpy as np
from numpy import loadtxt
import matplotlib.pyplot as plt
#import seaborn as sns
from pylab import *

file1='2019-logT375.txt'
file2='Bec-H-reos-T5mil.txt'
file3='S-5mil-H.txt'
file4='2019-logT4.txt'
file5='Bec-H-reos-T10mil.txt'
file6='scvh-T402.txt'

chab5mil = loadtxt(file1, usecols=[1,2])
beck5mil = loadtxt(file2, usecols=[0,2]) 
sau5mil = loadtxt(file3, usecols=[0,3]) #C1=densidad C3=presión
chab10mil = loadtxt(file4, usecols=[1,2])
beck10mil = loadtxt(file5, usecols=[0,2]) 
sau10mil = loadtxt(file6, usecols=[0,3]) #C1=densidad C3=presión
#print(chab5mil)
#print(beck5mil)
#print(sau5mil)
"""cambio de lugar las columnas"""
chab5mil[:,[0,1]] = chab5mil[:,[1,0]] 
chab10mil[:,[0,1]] = chab10mil[:,[1,0]] 
#beck5mil[:,[0,1]] = beck5mil[:,[1,0]] no cambiar
sau5mil[:,[0,1]] = sau5mil[:,[1,0]]
sau10mil[:,[0,1]] = sau10mil[:,[1,0]]

"""asignación de columnas a variables"""
den = chab5mil[:,0]
pres = chab5mil[:,1]
den2 = beck5mil[:,0]
pres2 = beck5mil[:,1]
den3 = sau5mil[:,0]
pres3 = sau5mil[:,1]

den4=chab10mil[:,0]
pres4=chab10mil[:,1]
den5 =beck10mil[:,0]
pres5 = beck10mil[:,1]
den6 = sau10mil[:,0]
pres6 = sau10mil[:,1]

fig1= plt.figure(figsize=(10,8),dpi=100)
axes=fig1.add_axes([0.4,0.5,1,1])

plt.subplot(211)
plt.plot(den,pres, '.', color='blue', label='Chabrier-2019')
plt.plot(den2,pres2, '+',color='red', label='Becker-2014')
plt.plot(den3,pres3, '+',color='black', label ='Saumon-1995')
plt.xlabel(r'log $\rho$ [$g/cm^3$]',fontsize=14)
plt.ylabel(r'Presión [$GPa$]',fontsize=14);
plt.grid()
plt.legend(loc='upper left')
plt.title("EOS-Hidrógeno")
#matplotlib.rcParams.update({'font.size': 10, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
plt.annotate("T = 5000 K", xy=(-0.9,5.5))

plt.subplot(212)
plt.plot(den4,pres4, '.', color='blue', label='Chabrier-2019')
plt.plot(den5,pres5, '+',color='red', label='Becker-2014')
plt.plot(den6,pres6-10, '+',color='black', label ='Saumon-1995')
plt.xlabel(r'log $\rho$ [$g/cm^3$]',fontsize=14);
plt.grid()
plt.legend(loc='upper left')
plt.annotate("T = 10000 K", xy=(-1.0,5.0))
#matplotlib.rcParams.update({'font.size': 10, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

                                   
plt.show()
fig1.savefig("EOS-H-5-10mil.png")


```
